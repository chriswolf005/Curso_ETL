```python
import requests
import pandas as pd

url = "https://pokeapi.co/api/v2/pokemon/charizard"
response = requests.get(url)
data = response.json()
print(data)
# Extraer campos relevantes y crear un DataFrame
df = pd.DataFrame([{
	'Nombre': data['name'].title(),
	'Altura': data['height'],
	'Peso': data['weight'],
	'Experiencia Base': data['base_experience']
}])
print(df)

```

```python
import requests

# Tu ciudad y API Key
city = 'London'
api_key = '0b2ae516cc4fbbfeedcde811cec49dbb'

# URL con unidades en grados Celsius
url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'

# Solicitud a la API
response = requests.get(url)
data = response.json()

# Extraer información útil
if response.status_code == 200:
    name = data['name']
    country = data['sys']['country']
    weather = data['weather'][0]['description'].title()
    temp = data['main']['temp']
    feels_like = data['main']['feels_like']
    humidity = data['main']['humidity']

    print(f"🌍 Ciudad: {name}, {country}")
    print(f"🌤️ Clima: {weather}")
    print(f"🌡️ Temperatura: {temp}°C")
    print(f"🤗 Sensación térmica: {feels_like}°C")
    print(f"💧 Humedad: {humidity}%")
else:
    print("❌ Error al obtener los datos:", data.get("message", "Error desconocido"))
```
